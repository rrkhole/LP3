```python
df['Geography'][df['Geography']=='France']=0
df['Geography'][df['Geography']=='Spain']=1
df['Geography'][df['Geography']=='Germany']=2 
df['Gender'][df['Gender']=='Male']=0
df['Gender'][df['Gender']=='Female']=1
```

```python
df.head()
```

```python
X_train=df.drop(columns='Exited')
Y_train=df['Exited']
```

```python
help(train_test_split)
```

```python
X_train1,X_test,Y_train1,Y_test=train_test_split(X_train,Y_train,test_size=0.2,shuffle=True,stratify=Y_train)
```

```python
#X=StandardScaler().fit_transform(X_train1)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
cls=MLPClassifier(random_state=1,max_iter=300)
```

```python
cls.fit(X_train1,Y_train1)
```

```python
Y_pred=cls.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
a=accuracy_score(Y_test,Y_pred)
```
