```python
curr_x=2
prev_step=1
precision=0.000001
iters=1 
max_iters=1000
df=lambda x:2*(x+3)
rate=0.01
while prev_step>precision and iters<max_iters:
    prev_x=curr_x
    curr_x=curr_x-rate*df(curr_x)
    prev_step=abs(curr_x-prev_x)
    iters=iters+1
    print("Iteration:",iters,"X:",curr_x)
print("Local minima at X=",curr_x)
```
